## Investigation Overview

- To view the bike rides of different hour and day of week
- To compare the rides of ebikes and docked-bikes
- To compare the rides of members and customers

## Dataset Overview

- May 2020 Data of Bay Wheels Trip service in Bay area, USA
- Main focus of the data includes: bike types, hour and day of ride,  and user types
- Data covered the registration of rides in 430 stations

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('clean_data.csv')
```

## Rides of days of week

- More rides during weekends than weekdays
- Sunday is the busiest time

```python
# plotting the bike rides based on the start day of week
plt.figure(figsize=(10,5))
base_color=sb.color_palette()[0]
orders = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
sb.countplot(data=df_clean, x='start_weekday_name', color=base_color, order=orders)
plt.title('Bike rides based on the day of week in May 2020')
plt.xlabel('Day of week')
plt.ylabel('Bike rides');
```

## Members vs. Customers for days of a week

- Both members and customers generally ride more during weekends than weekdays

- Members ride more than customers during weekdays while customers ride more during weekends

```python
# plotting ride counts of different user type based on the start day of week
plt.figure(figsize=[10,5])
sb.countplot(data=df_clean, x='start_weekday_name',hue='member_casual', order=orders)
plt.legend(['Customer', 'Member'], title='User type', loc='lower right', framealpha=1) # set legend at lower right with no transparency
plt.xlabel('Start day of week')
plt.ylabel('Count of rides')
plt.title('Rides of different user types based on day of week');
```

## Member vs Customers for different time of a day

- The peak hour of rides is 17 o'clock for both members and customers
- For both groups of users, the most silent time is midnight ( 3 - 4 o'clock)
- More rides from members than customers during morning time (4 - 9 o'clock)
- Customers are more active than members during day time (10 - 18 o'clock)

```python
# plotting ride counts of different user type based on the start day of week
plt.figure(figsize=[15,5])
sb.countplot(data=df_clean, x='start_hour',hue='member_casual')
plt.xlabel('Start hour of week')
plt.ylabel('Count of rides')
plt.title('Rides of different user types based on start hour')
plt.legend(['Customer','Member'], title='User type');
```

## Ebike vs. Docked-bike for days of a week


- More rides during weekends than weekdays for both types
- Saturday is the busiest day for both types
- More rides of docked-bikes than ebikes every day

```python
# plotting ride counts of different user type based on the start day of week
plt.figure(figsize=[10,5])
sb.countplot(data=df_clean, x='start_weekday_name',hue='rideable_type', order=orders)
plt.legend(['Ebike', 'Docked bike'], title='Bike types', loc='lower right', framealpha=1) # set legend at lower right with no transparency
plt.xlabel('Start day of week')
plt.ylabel('Count of rides')
plt.title('Rides for different bike types based on day of week');
```

## Ebike vs. Docked-bike for days of a week

- The peak hour of rides is 17 o'clock for both types
- The most silent time of rides is midnight for both types
- The rides of ebikes is close to or more than that of docked-bikes during early morning (6 - 8 o'clock)

```python
# plotting ride counts of different user type based on the start day of week
plt.figure(figsize=[15,5])
sb.countplot(data=df_clean, x='start_hour',hue='rideable_type')
plt.xlabel('Start hour of week')
plt.ylabel('Count of rides')
plt.title('Ride for different bike types based on start hour')
plt.legend(['Electric bike','Docked bike'], title='Bike type');
```
